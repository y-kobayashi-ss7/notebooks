```python
#データファイルの読み込み
import os
import struct
import numpy as np

def load_mnist(path, kind='train'):
    
    labels_path = os.path.join(path, '%s-labels-idx1-ubyte' % kind)
    images_path = os.path.join(path, '%s-images-idx3-ubyte' % kind)
    
    with open(labels_path, 'rb') as lbpath:
        
        magic, n = struct.unpack('>II', lbpath.read(8))
        labels = np.fromfile(lbpath, dtype=np.uint8)
        
    with open(images_path, 'rb') as imgpath:
        
        magic, num, rows, cols = struct.unpack(">IIII", imgpath.read(16))
        images = np.fromfile(imgpath, dtype=np.uint8).reshape(len(labels), 784)
        images = ((images/255.0) - 0.5) *2
        
    return images, labels
```

```python
import pandas as pd

path = 'mnist'
images, labels = load_mnist(path)

df = pd.DataFrame(images).iloc[1]
array = np.array(df)
m = np.reshape(array, (28,28))

pd.DataFrame(m)
```

```python
path = 'mnist'
X_train, y_train = load_mnist(path, kind='train')
X_test, y_test = load_mnist(path, kind='t10k')
```

```python
#サンプルデータの確認1
import matplotlib.pyplot as plt

fig,ax =plt.subplots(nrows=2, ncols=5, sharex=True, sharey=True)
ax = ax.flatten() #2×5を10に変換（for をそのまま回せる）
for i in range(10):
    img = X_train[y_train == i][0].reshape(28,28)
    ax[i].imshow(img, cmap='Greys')
    
ax[0].set_xticks([])
ax[0].set_yticks([])
plt.tight_layout()
plt.show()
```

```python
#読み込みデータの保存
import numpy as np

np.savez_compressed('mnist_scales.npz', X_train=X_train, y_train=y_train, X_test=X_test, y_test=y_test)
```
